### Plot

```python
def plot_decision_boundary(xs, model, colors=None, num=600):
    try:
        cmap = ListedColormap(colors.values())
    except AttributeError:
        cmap = None

    # Prepare the data to plot the decision boundary
    # Need to assign a color to each point in the mesh
    # [x1_min, x1_max] x [x2_min, x2_max]

    x_min, x_max = xs[:, 0].min() - 1, xs[:, 0].max() + 1
    y_min, y_max = xs[:, 1].min() - 1, xs[:, 1].max() + 1

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, num=num), np.linspace(y_min, y_max, num=num)
    )

    zz = model.predict(np.c_[xx.ravel(), yy.ravel()])
    zz = zz.reshape(xx.shape)
    # print(zz.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(
        xx,
        yy,
        zz,
        alpha=0.4,
        cmap=cmap,
        antialiased=True,
    )
```

```python
def plot_points(xs, ys, ft_names=(None, None), target_names=None, colors=None):
    if target_names is None:
        target_names = {}
    if colors is None:
        colors = {}

    for y_val in set(ys):
        plt.scatter(
            xs[ys == y_val, 0],
            xs[ys == y_val, 1],
            color=colors.get(y_val),
            alpha=0.8,
            label=target_names.get(y_val),
            edgecolor="black",
        )
    plt.legend(loc="best", shadow=False, scatterpoints=1)
    plt.title("3-Class classification")
    try:
        plt.xlabel(ft_names[0])
        plt.ylabel(ft_names[1])
    except TypeError:
        pass
```

```python
COLORS = {0: "green", 1: "red", 2: "blue"}

target_names = {k: v for k, v in enumerate(iris.target_names)}
target_names
```

```python
plot_points(X_test, y_test, features_2d, target_names, COLORS)
```

```python
plot_decision_boundary(X_test, model, COLORS)
```

```python
plot_decision_boundary(X_test, model, COLORS)
plot_points(X_test, y_test, features_2d, target_names, COLORS)
```

```python
accuracy_per_model = {"kNN": metrics.accuracy_score(y_test, y_pred)}
```

### Logisitc Regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy_per_model["Logistic Regression"] = metrics.accuracy_score(y_test, y_pred)
print(metrics.accuracy_score(y_test, y_pred))

plot_decision_boundary(X_test, model, COLORS)
plot_points(X_test, y_test, features_2d, target_names, COLORS)
```

### Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy_per_model["Decision Tree"] = metrics.accuracy_score(y_test, y_pred)
print(metrics.accuracy_score(y_test, y_pred))

plot_decision_boundary(X_test, model, COLORS)
plot_points(X_test, y_test, features_2d, target_names, COLORS)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy_per_model["Random Forest"] = metrics.accuracy_score(y_test, y_pred)
print(metrics.accuracy_score(y_test, y_pred))

plot_decision_boundary(X_test, model, COLORS)
plot_points(X_test, y_test, features_2d, target_names, COLORS)
```

### SVN

```python
from sklearn.svm import SVC

model = SVC()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy_per_model["SVM"] = metrics.accuracy_score(y_test, y_pred)
print(metrics.accuracy_score(y_test, y_pred))

plot_decision_boundary(X_test, model, COLORS)
plot_points(X_test, y_test, features_2d, target_names, COLORS)
```

### Comparison of the different models

```python
print("Accuracy per model:")
for model_name, accuracy in accuracy_per_model.items():
    print(f"{model_name}: {accuracy:.2f}")
```

La régression logistique domine avec une précision de 0.82, suivi des SVM à 0.80. Les forêts aléatoires suivent avec des précisions de 0.78 puis les k-NN à 0.76 et enfin les arbres de décision à 0.67. Cela suggère que les modèles linéaires et à marge maximale sont plus adaptés à ce jeu de données, tandis que les arbres de décision peuvent être moins performants en raison de leur tendance à surajuster.

## Exercise 2

* Apply a [PCA](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html) to all descriptors (there are 4) of the Iris dataset

* Give the ratio of the variance explained by each of the selected components, as well as their cumulative sum

* Apply one of the previous classifiers to the first two principal components

* Compare the quality of the predictions between those obtained with the first two principal components and those obtained with the first 2 descriptors (Exercise 1)

<span style="color:red">Implement your solution below</span>

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca.fit(iris.data)
pca_components = pca.transform(iris.data)

print("Variance ratios of the PCA components:", pca.explained_variance_ratio_)

models = [LogisticRegression(), KNeighborsClassifier(n_neighbors=3), DecisionTreeClassifier(), RandomForestClassifier(), SVC()]
for model in models:
  model.fit(pca_components, iris.target)
  y_pred = model.predict(pca_components)
  accuracy = metrics.accuracy_score(iris.target, y_pred)
  print(f"{model.__class__.__name__}: {accuracy:.2f}")
```

Les performances d'explications des différents modèles sont bien bien meilleures. Cela s'explique par le fait que les composantes principales capturent mieux la variance des données, permettant aux modèles de mieux séparer les classes. Choisir arbitrairement les deux premières composantes ne réflectais pas aussi bien la structure sous-jacente des données que la PCA.
